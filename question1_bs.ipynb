```python
import numpy as np
from scipy.stats import norm
import json
import csv


def d1_d2(S, K, t, T, sigma, r):
    dt = T - t
    d_1 = (np.log(S / K) + (r + sigma**2 / 2) * dt) / (sigma * np.sqrt(dt))
    d_2 = d_1 - sigma * np.sqrt(dt)
    return dt, d_1, d_2


def call_value_bs(S, K, t, T, sigma, r):
    dt, d_1, d_2 = d1_d2(S, K, t, T, sigma, r)
    return norm.cdf(d_1) * S - norm.cdf(d_2) * K * np.exp(-r * dt)


def put_value_bs(S, K, t, T, sigma, r):
    dt, d_1, d_2 = d1_d2(S, K, t, T, sigma, r)
    return norm.cdf(-d_2) * K * np.exp(-r * dt) - norm.cdf(-d_1) * S


columns = ["S", "K", "t", "T", "sigma", "r"]
values = [
    [100, 100, 0, 1, 0.3, 0.03],
    [100, 110, 0, 1, 0.3, 0.03],
    [100, 100, 0, 1.5, 0.3, 0.03],
    [100, 100, 0, 1, 0.4, 0.03],
    [100, 100, 0, 1, 0.3, 0.05],
    [110, 100, 0, 1, 0.3, 0.03],
]
ret = []
for idx, value in enumerate(values):
    inputs = dict(zip(columns, value))
    outputs = {}
    outputs["index"] = f"1.{idx+1}"
    outputs.update(inputs)
    outputs["call_value_bs"] = call_value_bs(**inputs)
    outputs["put_value_bs"] = put_value_bs(**inputs)
    ret.append(outputs)
with open("./result/q1/prices.json", "w", encoding="utf-8") as f:
    json.dump(ret, f, indent=4)
with open("./result/q1/prices.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=ret[0].keys())
    writer.writeheader()
    writer.writerows(ret)
```
