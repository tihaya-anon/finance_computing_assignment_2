prove

```math
    \mathbb{E}(XZ) = \mathbb{E}(\rho X^2 + \sqrt{1 - \rho^2}XY) \\
    = \rho \mathbb{E}(X^2) + \sqrt{1 - \rho^2} \mathbb{E}(XY) \\
    = \rho \cdot 1 + \sqrt{1 - \rho^2} \cdot 0 \\
    = \rho
```


````python
import numpy as np

N_SAMPLES = 200
RHO = 0.5
x_series = [np.random.normal() for _ in range(N_SAMPLES)]
y_series = [np.random.normal() for _ in range(N_SAMPLES)]
z_series = [RHO * x + np.sqrt(1 - RHO * RHO) * y for x, y in zip(x_series, y_series)]
x_mean = sum(x_series) / len(x_series)
z_mean = sum(z_series) / len(z_series)
x_var = sum([(x - x_mean) ** 2 for x in x_series])
z_var = sum([(z - z_mean) ** 2 for z in z_series])
xz_cov = sum([(x - x_mean) * (z - z_mean) for x, z in zip(x_series, z_series)])
correlation = xz_cov / np.sqrt(x_var * z_var)
print(correlation, RHO)
````
